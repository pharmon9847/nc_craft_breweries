```python
import pandas as pd
```

```python
# import data
beer_list_df = pd.read_csv("data/beer_list_df.csv")
```

```python
# isolate beer_style
beer_style = beer_list_df["beer_style"]
```

```python
master_style = []
style_type = ""

for style in beer_style:
    if style.endswith("India Pale Ale"):
        style_type = "IPA"
        master_style.append(style_type)
    elif style.endswith("Pale Ale"):
        style_type = "Pale Ale"
        master_style.append(style_type)
    elif style.endswith("Ale"):
        style_type = "Ale"
        master_style.append(style_type)
    elif style.endswith("Stout"):
        style_type = "Stout"
        master_style.append(style_type)
    elif style.endswith("Lager"):
        style_type = "Lager"
        master_style.append(style_type)
    elif style.endswith("Saison"):
        style_type = "Saison"
        master_style.append(style_type)
    elif style.endswith("Porter"):
        style_type = "Porter"
        master_style.append(style_type)
    elif style.endswith("Wheat Beer"):
        style_type = "Wheat Beer"
        master_style.append(style_type)
    elif style.endswith("Kolsch"):
        style_type = "Kolsch"
        master_style.append(style_type)
    elif style.endswith("Hefeweizen / Hefeweissbier"):
        style_type = "Hefeweizen"
        master_style.append(style_type)
    else:
        master_style.append("Other")
        
print(len(master_style))
```

```python
beer_list_df["master_style"] = master_style
```
