```python
import requests
import pandas as pd
import json
from config import api_key
```

```python
# http://api.brewerydb.com/v2/{endpoint}/?key=abcdef
# http://api.brewerydb.com/v2/locations?key=KEY&locality=denver&region=colorado

# results variable
total_results = []

# for loop to pull all page results 
for page_num in range(1,10):
    url = (f'http://api.brewerydb.com/v2/locations/?key={api_key}&region=NC&p={page_num}')
    print(f'Downloading page {page_num}')
    response = requests.get(url).json()
    data = response["data"]
    total_results = total_results + data
    


print(len(total_results))
```

```python
# Set up variables 
breweries = []
in_business = []
city = []
state = []
latitude = []
longitude = []
brewery_id = []
brewery_type = []
website = []
phone = []

for result in total_results:
    try: 
        breweries.append(result["brewery"]["name"])
    except (KeyError):
        breweries.append("NaN")
    try:     
        in_business.append(result["brewery"]["isInBusiness"])
    except (KeyError):
        in_business.append("NaN")
    try:
        city.append(result["locality"])
    except (KeyError):
        city.append("NaN")
    try:
        state.append(result["region"])
    except (KeyError):
        state.append("NaN")
    try:
        latitude.append(result["latitude"])
    except (KeyError):
        latitude.append("NaN")
    try:
        longitude.append(result["longitude"])
    except (KeyError):
        longitude.append("NaN")
    try:
        brewery_id.append(result["breweryId"])
    except (KeyError):
        brewery_id.append("NaN")
    try:
        brewery_type.append(result['locationTypeDisplay'])
    except (KeyError):
        brewery_type.append("NaN")
    try:
        website.append(result["website"])
    except (KeyError):
        website.append("NaN")
    try:
        phone.append(result["phone"])
    except (KeyError):
        phone.append("NaN")
```

```python
columns = ["brewery_id","breweries", "in_business", "city", "state", "latitude", "longitude", "brewery_type", "phone"]

nc_breweries = pd.DataFrame(columns=columns)

nc_breweries["brewery_id"] = brewery_id
nc_breweries["breweries"] = breweries
nc_breweries["in_business"] = in_business
nc_breweries["city"] = city
nc_breweries["state"] = state
nc_breweries["latitude"] = latitude
nc_breweries["longitude"] = longitude
nc_breweries["brewery_type"] = brewery_type
nc_breweries["website"] = website
nc_breweries["phone"] = phone

nc_breweries
```

```python
in_business = nc_breweries.loc[nc_breweries["in_business"] == "Y", :]
in_business
```

```python
nc_breweries_df = in_business.reset_index()
```

```python
del nc_breweries_df["index"]
```
