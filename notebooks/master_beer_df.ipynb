```python
import pandas as pd
```

```python
nc_breweries_df = pd.read_csv("data/nc_breweries_df.csv")
beer_list_df = pd.read_csv("data/beer_list_df.csv")
```

```python
# create list of brewery id's that returned a beer list 
brewery_id = beer_list_df["brewery_id"]
```

```python
# remove duplicates 
cleaned_brewery_id = []

for brewery in brewery_id:
    if brewery not in cleaned_brewery_id:
        cleaned_brewery_id.append(brewery)
```

```python
# remove breweries that did not return a beer list 
for index, row in nc_breweries_df.iterrows():
    if row[0] not in cleaned_brewery_id:
        nc_breweries_df.drop(index, inplace=True)
```

```python
nc_breweries_df.to_csv("data/nc_breweries_df.csv")
```

```python
# merge the two csv's 
master_beer_df = pd.DataFrame.merge(nc_breweries_df, beer_list_df, on="brewery_id")
```
