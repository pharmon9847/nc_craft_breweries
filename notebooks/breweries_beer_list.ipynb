```python
import pandas as pd
import requests 
import json
from config import api_key
```

```python
# import nc_breweries_df to isolate brewery_id
nc_breweries_df = pd.read_csv("data/nc_breweries_df.csv")
```

```python
# set the brewery ids to a list 
brewery_id_list = nc_breweries_df["brewery_id"]
```

```python
# remove duplicates from the brewery_id_list

brewery_id = []

for brewery in brewery_id_list:
    if brewery not in brewery_id:
        brewery_id.append(brewery)

# check length
print(len(brewery_id))
```

```python
# split into two lists due to being limited to 200 api calls/day 
brewery_id_1 = brewery_id[0:176]
brewery_id_2 = brewery_id[176:]
```

```python
# Set up a for loop to perform the api calls 

total_beer_results = []
total_breweries_checked = 0
no_beer_list = []

# Day 1
for brewery in brewery_id_1:
    url = (f'http://api.brewerydb.com/v2/brewery/{brewery}/beers/?key={api_key}&withBreweries=Y')
    print(brewery + " beer list is being downloaded")
    response = requests.get(url).json()
    try:
        data = response["data"]
    except (KeyError):
        print(brewery + " does not have a beer list")
        no_beer_list.append(brewery)
    total_beer_results = total_beer_results + data
    total_breweries_checked = total_breweries_checked + 1

print(f'{total_breweries_checked} breweries beer lists were checked')
print(len(total_beer_results))
```

```python
# Day 2
for brewery in brewery_id_2:
    url = (f'http://api.brewerydb.com/v2/brewery/{brewery}/beers/?key={api_key}&withBreweries=Y')
    print(brewery + " beer list is being downloaded")
    response = requests.get(url).json()
    try:
        data = response["data"]
    except (KeyError):
        print(brewery + " does not have a beer list")
        no_beer_list.append(brewery)
    total_beer_results = total_beer_results + data
    total_breweries_checked = total_breweries_checked + 1

print(f'{total_breweries_checked} breweries beer lists were checked')
print(len(total_beer_results))
```

```python
#set up empty variables 
abv = []
availability = []
brewery_id = []
ibu = []
beer_name = []
beer_id = []
beer_style = []
is_retired = []

for result in total_beer_results:
    try: 
        abv.append(result['abv'])
    except (KeyError):
        abv.append("NaN")
    try:
        availability.append(result["available"]["name"])
    except (KeyError):
        availability.append("NaN")
    try: 
        brewery_id.append(result["breweries"][0]["id"])
    except (KeyError):
        brewery_id.append("NaN")
    try:
        ibu.append(result["ibu"])
    except (KeyError):
        ibu.append("NaN")
    try:
        beer_name.append(result["name"])
    except (KeyError):
        beer_name.append("NaN")
    try:
        beer_id.append(result["id"])
    except (KeyError):
        beer_id.append("NaN")
    try:
        beer_style.append(result["style"]["name"])
    except (KeyError):
        beer_style.append("NaN")
    try:
        is_retired.append(result["isRetired"])
    except (KeyError):
        is_retired.append("NaN")
```

```python
columns = ["brewery_id", "beer_id", "availability", "beer_name", "beer_style", "abv", "ibu", "is_retired"]

nc_breweries_beer_list = pd.DataFrame(columns=columns)

nc_breweries_beer_list["brewery_id"] = brewery_id
nc_breweries_beer_list["beer_id"] = beer_id
nc_breweries_beer_list["availability"] = availability
nc_breweries_beer_list["beer_name"] = beer_name
nc_breweries_beer_list["beer_style"] = beer_style
nc_breweries_beer_list["abv"] = abv
nc_breweries_beer_list["ibu"] = ibu
nc_breweries_beer_list["is_retired"] = is_retired

nc_breweries_beer_list
```

```python
# clean out the retired beers
cleaned_beer_list = nc_breweries_beer_list.loc[nc_breweries_beer_list["is_retired"] == "N", :] 
```

```python
beer_list_df = cleaned_beer_list.reset_index()
beer_list_df
```

```python
del beer_list_df["index"]
```

```python
beer_list_df.to_csv("data/beer_list_df.csv")
```
